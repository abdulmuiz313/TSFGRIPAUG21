```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading Dataset
df=pd.read_csv("globalterrorismdb.csv",encoding='ISO-8859-1')
```

```python
#Renaming columns,as this is a raw dataset containing some improper names,which is difficult to understand
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                   'region_txt':'Regions','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                     'nwound': 'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type','motive':'Motive'
                    },inplace=True)
```

```python
#copying required data into the same data frame(overwriting)
df=df[['Year','Month','Day','Country','Regions','state','city','latitude','longitude','AttackType','Killed','Wounded','Target',
        'Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
df
```

```python
#checking for null values in the new dataset
df.isnull().sum()
```

# Some raw inferences from data are as follows:

```python
print("Country with most attacks:",df['Country'].value_counts().idxmax())
print("Region with most attacks:",df['Regions'].value_counts().idxmax())
print("Year with most attacks:\t",df['Year'].value_counts().index[0])
print("City with most attacks:\t",df['city'].value_counts().index[1])
print("Group with most attacks:",df['Group'].value_counts().index[1])
print("Mostly used type of attack:",df['AttackType'].value_counts().idxmax())
print("State with most attacks:",df['state'].value_counts().iloc[:1])
```

```python
#calculating total number of attacks per year
X=df['Year'].unique()
Y=df['Year'].value_counts(dropna=True).sort_index()
```

```python
#Plot of Attacks per decade.
plt.figure(figsize=(18,10))
sns.lineplot(x=X,y=Y)#,palette='Blues')
plt.xticks(rotation = 90)
plt.xlabel('Attack Year')
plt.ylabel('Attack Counts')
plt.title('Attacks per decade',fontsize=16)
plt.show()
```

```python
#Number of attacks per year
plt.subplots(figsize=(15,6))
sns.countplot('Year',data = df,palette='Blues',edgecolor='Black')
plt.xticks(rotation=90)
plt.title('Number of Terrorist Activites Each Year',fontsize=16)
plt.xlabel("Years",fontsize=14)
plt.ylabel("Attacks Count",fontsize=14)
plt.show()
```

```python
df['Wounded']=df['Wounded'].fillna(0).astype(int)
df['Killed'] = df['Killed'].fillna(0).astype(int)
df['casualities'] = df['Killed'] + df['Wounded']
```

```python
df = df.sort_values(by='casualities',ascending=False)[:40]
```

```python
#Creating a pivot table for casualties count
heat = df.pivot_table(index = 'Country',columns='Year',values = 'casualities')
heat.fillna(0,inplace=True)
```

```python
heat.head()
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
heatmap = go.Heatmap(z=heat.values, x=heat.columns, y=heat.index, colorscale='Blues')
data2 = [heatmap]
layout = go.Layout(
    title='Top 40 Worst Terror Attacks in History from 1982 to 2017',
    xaxis = dict(ticks='', nticks=20),
    yaxis = dict(ticks='')
)
fig = go.Figure(data=data2, layout=layout)
py.iplot(fig, filename='heatmap',show_link=False)
```

```python
#Total number of people killed
Killed = df.loc[:,'Killed']
print('Number of people killed by terror attack:',int(sum(Killed.dropna())))
```

```python
#seperating by type of attack
attack = df.loc[:,'AttackType']
typeKilled = pd.concat([attack,Killed],axis=1) 
```

```python
typekilled=typeKilled.reset_index(drop=True)
```

```python
typekilled.head()
```

```python
#creating 2d pivot table
typeKilledData = typeKilled.pivot_table(columns='AttackType',values = 'Killed',aggfunc = 'sum')
typeKilledData
```

```python
typeKilledData.info()
```

```python
countrydata = df.loc[:,'Country'] 
kill_per_country = pd.concat([countrydata,Killed],axis=1)
```

```python
kill_per_country = kill_per_country.pivot_table(columns = 'Country', values='Killed',aggfunc='sum')
kill_per_country
```

```python
df['Group'].value_counts()[1:16].plot(kind='pie',figsize=(9,9),autopct='%.2f%%',fontsize=12)
```

```python
#Setting Figure Size
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 25
fig_size[1] = 25
plt.rcParams["figure.figsize"] = fig_size
```

```python
labels = kill_per_country.columns.tolist()
labels = labels[:50]
index = np.arange(len(labels))
transpose = kill_per_country.T
values = transpose.values.tolist()
values = values[:50]
values = [int(i[0]) for i in values]
colors = ['red','green','purple','yellow','brown','black','gray','magenta','orange']
fig,ax = plt.subplots(1,1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 25
fig_size[1] = 25
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index,values,color = colors,width=0.9)
plt.xlabel('Countries',fontsize=20)
plt.ylabel('Killed People',fontsize = 20)
plt.xticks(index,labels,fontsize=18,rotation=90)
plt.yticks(fontsize=18)
plt.title('Number of people killed by countries',fontsize = 20)
plt.show()
```

```python
#monthwise attacks visualization
sns.barplot(x = df["Month"].value_counts()[:20].index, y=df["Month"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", fontsize=18)
plt.xticks(rotation=90,fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel("Attack month", fontsize=18)
plt.title(" Attacks BY MONTH ", size=25, fontweight="bold")
```

```python
#creating cross table using 2 columns and plotting Terrorist Activites by region in each year
pd.crosstab(df.Year, df.Regions).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('No. of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
#top 10 most cities affected
df['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='orange')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Cities",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20)
plt.show()
```

```python
#Attacks by group(top 20 groups)
sns.barplot(x = df["Group"].value_counts()[:20].index, y=df["Group"].value_counts()[:20].values)
plt.ylabel("Total Attacks", fontsize=18)
plt.xticks(rotation=90,fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel("Attacking Group", fontsize=18)
plt.title("Attacks by Group", size=25, fontweight="bold")
plt.show()
```
